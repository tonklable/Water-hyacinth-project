```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

img = cv.imread('./1x/Asset 1.png',cv.IMREAD_GRAYSCALE)
img=(img-121)//(138-121)
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if img[i,j]==7:
            img[i,j]=16
plt.hist(img[:,:].ravel(),255,[1,15],color='red')
plt.figure(figsize=(20,16),dpi=100)
plt.imshow(img, cmap="gray")
plt.grid(False)
print(img[:,:].ravel())
print(-10//5)
unique, counts = np.unique(img, return_counts=True)
a=dict(zip(unique, counts))
b=a[10]+a[11]+a[12]+a[13]+a[14]+a[15]
print(a)
print(img.shape[0]*img.shape[1]-a[16],b)
```

```python
#Calculate percentage of water hyacinth
r=img.shape[0]*img.shape[1]-a[16]
print("{0}{1}".format(b*100//r,"%"))
```
